```python
import pandas as pd
```

```python
dset=pd.read_csv("diabetes.csv")
```

```python
dset
```

```python
from sklearn.preprocessing import StandardScaler

kl=StandardScaler()

kl.fit(dset.drop("Outcome",axis=1))

stsc=kl.transform(dset.drop("Outcome",axis=1))

new_df=pd.DataFrame(stsc,columns=dset.columns[:-1])
```

```python
new_df
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(new_df,dset["Outcome"],test_size=0.3,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.neural_network import MLPClassifier
zh=MLPClassifier(random_state=1)
```

```python
zh = MLPClassifier()
zh.fit(x_train,y_train)

klm=zh.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(klm,y_test))
print(classification_report(klm,y_test))

```
